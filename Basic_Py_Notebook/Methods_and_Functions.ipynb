```python
# return true if any number is even inside a list 

def check_list_for_even_number(num_list):
    for number in num_list:
        if number % 2 == 0:
            return True
        else :
            pass
    return False


check_list_for_even_number([1,3,2])
```

```python
# return all the even number 

def check_list_for_even_number(num_list):
    even_list = []
    for number in num_list:
        if number % 2 == 0:
            even_list.append(number)
        else :
            pass
    return even_list

check_list_for_even_number([1,3,2,4,6])
```

```python
work_h = [("A",100),("B",10),("C",300),("D",800)]

def check_work(work):
    max_h = 0
    name = ''
    for w,n in work:
        if n > max_h:
            max_h = n
            name = w
        else:
            pass
        
    return(name,max_h)

check_work(work_h)
```

```python
def player_guess():
    
    guess = ''
    while guess not in ['0','1','2']:
        guess = input("Pick a Number 0 , 1 or 2")
        
    return int(guess)
```

```python
def check_guess(mylist,guess):
    if mylist[guess]== ["0"]:
        print("Correct")
    else:
        print("Wrong Gess")
        print(mylist)
```

```python
# **kwargs

def myfunc(**kwarg):
    print(kwarg)
    if 'fruit' in kwarg:
        print(f"My Fruit of choice is {kwarg['fruit']}")
    else :
        print("No Fruit")
        
        
myfunc(fruit="Mango")
```
